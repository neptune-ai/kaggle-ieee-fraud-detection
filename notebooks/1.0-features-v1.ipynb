```python
from collections import OrderedDict
from pandarallel import pandarallel
pandarallel.initialize()

def split_email(x, colname):
    if type(x) == float and np.isnan(x):
        email_first, email_rest = None, None
    else:
        split = x.split('.')
        email_first = split[0]
        email_rest = '.'.join(split[1:])
    return pd.Series(OrderedDict([('{}_first'.format(colname),email_first),
                                  ('{}_rest'.format(colname),email_rest)]))

train['P_emaildomain'].parallel_apply(lambda x: split_email(x, colname='P_emaildomain'))
train['R_emaildomain'].parallel_apply(lambda x: split_email(x, colname='R_emaildomain'))
```
